# Data Preprocessing Report
Data Loading and Data Processing

This report demonstrates the preprocessing of the shopping behavior dataset using Python.
The dataset is cleaned by handling missing values, correcting inconsistent data, encoding categorical variables, and normalizing numerical columns.

## Data Preprocessing

Before performing analysis and OLAP operations, the dataset was cleaned and preprocessed to ensure consistency, accuracy, and suitability for analysis. The following preprocessing steps were performed:

### 1. Dataset Loading and Initial Exploration
The dataset was loaded using the **Pandas** library. Initial exploration was performed to understand the structure of the dataset. Functions such as `head()`, `tail()`, `shape`, `info()`, and `describe()` were used to examine the data, identify data types, and observe statistical summaries of numerical columns.

### 2. Handling Missing Values
The dataset was checked for missing values using `isnull().sum()`. Missing values were handled based on the column type:
- **Numerical columns** such as Age and Purchase Amount were filled using the **mean value**.
- **Categorical columns** such as Item Purchased, Category, Subscription Status, and Frequency of Purchases were filled using the **mode (most frequent value)**.

This ensures that the dataset remains complete without losing important records.

### 3. Data Cleaning and Standardization
Some categorical columns contained inconsistent or incorrect values. These were standardized to maintain uniformity across the dataset.

Examples include:
- Variations of **Gender** were corrected and standardized (e.g., incorrect spellings were converted to "Male" or "Female").
- The **Season** column values were standardized to maintain consistent seasonal categories.
- The **Payment Method** column contained different variations that were standardized into consistent categories such as Credit Card, Debit Card, or Online Payment.

### 4. Removing Duplicate Records
Duplicate rows were identified using the `duplicated()` function and removed using `drop_duplicates()`. Removing duplicate records ensures that each transaction is counted only once and prevents biased analysis.

### 5. Encoding Categorical Variables
Some categorical columns such as **Gender, Season, and Payment Method** were converted into numerical format using categorical encoding. This transformation makes the data suitable for analysis and machine learning algorithms that require numerical input.

### 6. Data Normalization
The **Purchase Amount (USD)** column was normalized using **Min-Max Scaling** to scale values between 0 and 1. The formula used was:

Normalized Value = (Value − Minimum Value) / (Maximum Value − Minimum Value)

Normalization ensures that numerical features are on a comparable scale and prevents larger values from dominating the analysis.

### 7. Saving the Cleaned Dataset
After completing all preprocessing steps, the cleaned dataset was saved as a new file. This processed dataset was then used for further analysis, visualization, OLAP operations, and association rule mining.

```python
import pandas as pd
import numpy as np
df = pd.read_excel("shopping_behaviour.xlsx")
df.head()
df.tail()
df.shape
df.columns
df.dtypes
df.info()
df.isnull().sum()
df.duplicated().sum()
df.describe()
print(df["Gender"].unique())
print(df["Season"].unique())
print(df["Payment Method"].unique())
```

```python
# Fix Gender Column
df["Gender"] = df["Gender"].replace(
    ['Mle','Mael','M@Le','M#Le','M*Le','M4Le','M@L3','<Male>','Male😊'],
    'Male'
)

df["Gender"] = df["Gender"].fillna(df["Gender"].mode()[0])

# Fix Season Column
# -----------------------------
df["Season"] = df["Season"].replace("Autumn", "Fall")

# -----------------------------
# Fix Payment Method
# -----------------------------
df["Payment Method"] = df["Payment Method"].replace({
    "Credit": "Credit Card",
    "Debit": "Debit Card",
    "Visa": "Credit Card",
    "User@Email.Com": "Online Payment"
})

# -----------------------------
# Convert Purchase Amount to Numeric
# -----------------------------
df["Purchase Amount (USD)"] = pd.to_numeric(
    df["Purchase Amount (USD)"], errors='coerce'
)

# -----------------------------
# Handle Missing Values
# -----------------------------
df["Purchase Amount (USD)"] = df["Purchase Amount (USD)"].fillna(df["Purchase Amount (USD)"].mean())
df["Age"] = df["Age"].fillna(df["Age"].mean())

df["Item Purchased"] = df["Item Purchased"].fillna(df["Item Purchased"].mode()[0])
df["Category"] = df["Category"].fillna(df["Category"].mode()[0])
df["Subscription Status"] = df["Subscription Status"].fillna(df["Subscription Status"].mode()[0])
df["Frequency of Purchases"] = df["Frequency of Purchases"].fillna(df["Frequency of Purchases"].mode()[0])

# -----------------------------
# Remove Duplicates
# -----------------------------
df = df.drop_duplicates()

# -----------------------------
# Encode Categorical Columns
# -----------------------------
df["Gender"] = df["Gender"].astype("category").cat.codes
df["Season"] = df["Season"].astype("category").cat.codes
df["Payment Method"] = df["Payment Method"].astype("category").cat.codes

# -----------------------------
# Normalize Purchase Amount
# -----------------------------
df["Purchase Amount (USD)"] = (
    df["Purchase Amount (USD)"] - df["Purchase Amount (USD)"].min()
) / (
    df["Purchase Amount (USD)"].max() - df["Purchase Amount (USD)"].min()
)

# Show cleaned dataset
print(df.head())
print(df.info())

# Save cleaned dataset
df.to_csv("cleaned_shopping_dataset.csv", index=False)
print("dataset cleaned and saved as cleaned_shopping_dataset.csv")
```
