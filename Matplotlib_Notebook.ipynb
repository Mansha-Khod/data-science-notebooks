```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

<h1>First Plot</h1>

```python
x=np.arange(0,100,10)
y=np.arange(100,1,-10)
a=np.arange(12,112,10)
b=np.arange(112,12,-10)

plt.figure(figsize=(5,3),dpi=200)

plt.plot(x,y,label='2X',color="pink",linewidth=2,linestyle='-',marker='.',markersize=10,markeredgecolor='red')
# ## plt.plot(a,b,'b.-',label='3X') ANOTHER WAY TO MAKE A LINE

##SECOND LINE
plt.plot(b,a,'g',label='3X')
plt.title('My First Graph !!',fontdict={'fontname':'cursive','fontsize':20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks(np.arange(0,100,10))
plt.yticks(np.arange(100,1,-10))
plt.legend()
plt.savefig('MyFirstGraph.png',dpi=300)
plt.show()
```

## LINE DIAGRAM

```python
gas=pd.read_csv('gas_prices.csv')
plt.figure(figsize=(8,5))

plt.title('Gas Prices Over Time (in USD)',fontdict={'fontweight':'bold','fontsize':20})

plt.plot(gas.Year,gas.USA,label='USA',marker='.')
plt.plot(gas.Year,gas.Canada,label='Canada',marker='.',color='yellow')
plt.plot(gas.Year,gas['South Korea'],label='South Korea',marker='.',color='green')
plt.plot(gas.Year,gas.Australia,label='Australia',marker='.',color='red')

plt.xlabel('Year')
plt.ylabel('US Dollers')

plt.legend()
plt.xticks(gas.Year[0::2])
plt.show()
```

```python
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year,gas[country],label=country,marker='.')
plt.legend()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count().iloc[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count().iloc[0]
labels=["Left",'Right']

plt.title('Players Preferred Foot',fontdict={'fontweight':'bold','fontsize':15})
plt.pie([left,right],labels=labels,colors=['red','pink'],autopct='%.2f %%',wedgeprops={"linewidth": 1, "edgecolor": "black"});
```

```python
fifa.Weight=[int(x.strip('lbs')) if type(x)==str else x  for x in fifa.Weight]
plt.style.use('ggplot')
light = fifa.loc[fifa.Weight < 125].count().iloc[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count().iloc[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count().iloc[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count().iloc[0]
heavy = fifa[fifa.Weight >= 200].count().iloc[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
labels=['Under 125','125-150','150-175','175-200','Over 200']

plt.pie(weights,labels=labels,autopct='%.2f %%',pctdistance=0.8)
plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count().iloc[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count().iloc[0]
labels=["Left",'Right']

plt.title('Players Preferred Foot',fontdict={'fontweight':'bold','fontsize':15})
plt.pie([left,right],labels=labels,colors=['red','pink'],autopct='%.2f %%',wedgeprops={"linewidth": 1, "edgecolor": "black"});
```

```python
fifa.Weight=[int(x.strip('lbs')) if type(x)==str else x  for x in fifa.Weight]
plt.style.use('ggplot')
light = fifa.loc[fifa.Weight < 125].count().iloc[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count().iloc[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count().iloc[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count().iloc[0]
heavy = fifa[fifa.Weight >= 200].count().iloc[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
labels=['Under 125','125-150','150-175','175-200','Over 200']
explode=[0.6,0.1,0,0,0.6]

plt.title("Player Weights Pie Chart", loc='center', fontsize=14, fontweight='bold', color='navy')

plt.pie(weights,labels=labels,autopct='%.2f %%',pctdistance=0.8,explode=explode)
plt.show()
```

```python
plt.style.use('default')
plt.figure(figsize=(5,8))

plt.title('Proffesional Football Team Comparison')
plt.ylabel('Overall Rating')

barcolona=fifa.loc[fifa.Club=='FC Barcelona']['Overall']
madrid=fifa.loc[fifa.Club=='Real Madrid']['Overall']
revs=fifa.loc[fifa.Club=='New England Revolution']['Overall']
labels=['FC Barcelona','Real Madrid','NE Revolution']


boxes=plt.boxplot([barcolona,madrid,revs],labels=labels,patch_artist=True,medianprops={'linewidth':2,'color':'red'})

for box in boxes['boxes']:
    #set edge color
    box.set(color='#4286f4',linewidth=2)
    # change fill color
    box.set(facecolor='pink')
    


plt.show()
```
